Общие методы заполнения отсутствующих данных включают удаление, заполнение фиксированным значением, среднее/моду/медиану.

Мы можем использовать fillna() для заполнения отсутствующего значения фиксированным значением следующим образом:

```python
titanic_fillna = titanic.Cabin.fillna("Unknown")
titanic_fillna
```

Набор данных этого проекта невелик, поэтому нецелесообразно напрямую использовать фиксированное значение для заполнения
данные, и он не может отражать характеристики этой части отсутствующих данных. Поэтому мы используем
алгоритм случайного леса для прогнозирования отсутствующего значения данных с помощью другого атрибута
ценности.

```python
from sklearn.ensemble import RandomForestRegressor
def set_missing_age(df):
# We take the numeric features and put them into a random forest for training.
    age_df = df[['Age','Fare','Parch','SibSp','Pclass']]
# The passengers are divided into two parts: known age and unknown age
    known_age = age_df[age_df.Age.notnull()].values
    unknown_age = age_df[age_df.Age.isnull()].values
# Target data y
    y = known_age[:,0]
# Feature attribute data x
    x = known_age[:,1:]
# Using random forest algorithm for fitting
    rfr = RandomForestRegressor(random_state=0,n_estimators=2000,n_jobs=-1)
    rfr.fit(x,y)
# The trained model is used for prediction
    predictedAges = rfr.predict(unknown_age[:,1::])
# filling missing data
    df.loc[(df.Age.isnull()),'Age'] = predictedAges
    return df
# Fill in the missing value of age
titanic_preprocess = set_missing_age(titanic)
titanic_preprocess
```

Есть только две записи с пропущенными значениями атрибута Embarked, и их количество равно
маленький, поэтому мы сразу удаляем эти две записи

```python
titanic_preprocess = titanic_preprocess.drop(
    titanic_preprocess[titanic_preprocess.Embarked.isnull()].index)
titanic_preprocess.shape
```

Мы можем использовать функцию Duplicated() для обнаружения повторяющихся значений.
Используйте функцию drop_duplicates() для удаления повторяющихся значений.

```python
titanic_preprocess[titanic_preprocess.duplicated()].count()
```

# Step 5. Feature engineering

Разработка признаков — очень важная часть анализа данных и моделирования. В этой части на
с одной стороны, нам нужно выполнить обработку данных на основе предыдущего анализа (нормализованного,
уменьшение габаритов и др.); С другой стороны, нам нужно обрабатывать данные в соответствии с требованиями
модель, которую мы выбираем (факторизация, преобразование данных и т. д.).

Выбор признаков — еще один важный этап обработки в разработке признаков. Этот процесс выполняется
через весь процесс построения-прогнозирования-анализа модели, что является важным
этап оптимизации модели и требует непрерывной итерации и тестирования.

Категориальные признаки обычно являются не непрерывными значениями, а категориальными значениями. Например,
пол человека может быть "мужской", "женский". Когда мы используем моделирование логистической регрессии, необходимо
входные функции являются числовыми. Следовательно, нам нужно закодировать эти нечисловые функции в
числовые особенности.

Мы можем использовать реализацию функции get_dummies() pandas для быстрого кодирования категории oсобенности. Такие как:

```python
dummies_Cabin = pd.get_dummies(titanic_preprocess['Cabin'], prefix= 'Cabin')
dummies_Embarked = pd.get_dummies(titanic_preprocess['Embarked'], prefix= 'Embarked')
dummies_Sex = pd.get_dummies(titanic_preprocess['Sex'], prefix= 'Sex')
dummies_Pclass = pd.get_dummies(titanic_preprocess['Pclass'], prefix= 'Pclass')
dummies_Cabin.head()
```

```python
titanic_oneHot = pd.concat([titanic_preprocess, dummies_Cabin, dummies_Embarked, dummies_Sex, dummies_Pclass], axis = 1)
titanic_oneHot.drop(['Pclass', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
titanic_oneHot.head(10).T
```

Сводка способов избавиться от пустых категориальных значений [Тут](https://habr.com/ru/post/663414/)

Теперь мы успешно преобразовали все эти категориальные признаки в числовые признаки 0,1.

Это также можно сделать с помощью метода sklearn.preprocessing.OneHotEncoder() Scikit Learn.

Когда мы используем для моделирования такие алгоритмы, как логистическая регрессия и градиентный спуск, если
разрыв между значениями каждого атрибута слишком велик, это серьезно повлияет на конвергенцию
скорость! Он даже не сходится! Поэтому нам нужно проверить диапазон каждого числового значения.
сначала атрибут. Если диапазон большой,
мы должны нормализовать эти свойства перед моделированием.

Мы можем использовать метод ptp() pandas для вычисления диапазона числовых атрибутов. Мы рассчитываем
диапазон двух атрибутов Age и Fare следующим образом:

```python
print('Range of Fare: ',titanic_oneHot['Fare'].values.ptp())
print('Range of Age: ',titanic_oneHot['Age'].values.ptp())
```

Модуль предварительной обработки в пакете scikit-learn предоставляет такие методы, как
StandardScaler() и MinMaxScaler() для масштабирования объектов до указанного диапазона.

Теперь мы используем метод StandardScaler() для нормализации Age и Fare до диапазона [-1, 1].

```python
import sklearn.preprocessing as preprocessing
scaler = preprocessing.StandardScaler()
age_scale_param = scaler.fit(titanic_oneHot['Age'].values.reshape(-1,1))
titanic_oneHot['Age_scaled'] = scaler.fit_transform(titanic_oneHot['Age'].values.reshape(-1,1), age_scale_param)
fare_scale_param = scaler.fit(titanic_oneHot['Fare'].values.reshape(-1,1))
titanic_oneHot['Fare_scaled'] = scaler.fit_transform(titanic_oneHot['Fare'].values.reshape(-1,1), fare_scale_param)
titanic_oneHot.head(10)
```

```python
titanic_oneHot.drop(['Age','Fare','Age_group','Age_cut'], axis=1, inplace=True)
```

Модель логистической регрессии — это алгоритм бинарной классификации, основанный на линейной регрессии.
Выберем эту модель для моделирования.

```python
from sklearn import linear_model
import numpy as np
# Split the data set into training set and test set
data_train = titanic_oneHot[:int(len(titanic_oneHot) * 0.75)]
data_test = titanic_oneHot[int(len(titanic_oneHot) * 0.75):]
# Use regular expressions to extract the desired properties
train_df = data_train.filter(regex='Survived|Age_.*|SibSp|Parch|Fare_.*|Embarked_.*|Sex_.*|Pclass_.*|')
train_np = train_df.values
test_df = data_test.filter(regex='Age_.*|SibSp|Parch|Fare_.*|Embarked_.*|Sex_.*|Pclass_.*|')
# get y
y = train_np[:,0]
# get x
x = train_np[:,1:]
# Logistic regression was used to fit the trained
clf = linear_model.LogisticRegression(solver='liblinear', C=1.0,penalty='l2',tol=1e-6)
clf.fit(x,y)
# Predict with the test data
predictions = clf.predict(test_df)
result = pd.DataFrame({'PassengerId':data_test['PassengerId'].values,'Survived':predictions.astype(np.int32)})
print(result.head())
# The coefficient of each attribute in the model is used to analyze the contribution data
feature_corr = pd.DataFrame({'columns':list(train_df.columns)[1:],'coef':list(clf.coef_.T)})
print(feature_corr)
```
